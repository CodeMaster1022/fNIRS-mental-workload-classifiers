```python
Finetuning_OtherSubject64_df
```

### Get 97.5 and 2.5 percentile using bootstrapping

To generate 1 bootstrap sample, first sample the 71 subjects to include for this bootstrap sample.
Then for each included subject, sample the chunk to calculate accuracy for this subject

```python
num_bootstrap_samples=5000
upper_percentile=97.5
lower_percentile=2.5

candidate_subjects = np.array([1, 13, 14, 15, 20, 21, 22, 23, 24, 25, 27, 28, 29, 31, 32, 34,
 35, 36, 37, 38, 40, 42, 43, 44, 45, 46, 47, 48, 49,  5, 51, 52,
 54, 55, 56, 57, 58, 60, 61, 62, 63, 64, 65, 68, 69,  7, 70, 71,
 72, 73, 74, 75, 76, 78, 79, 80, 81, 82, 83, 84, 85, 86, 91, 92, 93,
 94, 95, 97])
```

```python
len(candidate_subjects)
```

#### OtherSubject64

```python
lookup_table = Finetuning_OtherSubject64_df.copy()
Finetuning_OtherSubject64_upper_percentile, Finetuning_OtherSubject64_lower_percentile = bootstrapping(candidate_subjects, lookup_table, num_bootstrap_samples=num_bootstrap_samples, upper_percentile=upper_percentile, lower_percentile=lower_percentile)
print('Finetuning_OtherSubject64:')
print('{} percentile is {}; {} percentile is {}'.format(upper_percentile, Finetuning_OtherSubject64_upper_percentile, lower_percentile, Finetuning_OtherSubject64_lower_percentile))
```

```python
print('{} percentile is {}; {} percentile is {}'.format(upper_percentile, Finetuning_OtherSubject64_upper_percentile, lower_percentile, Finetuning_OtherSubject64_lower_percentile))
```
